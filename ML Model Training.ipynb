### 1. Check for the outliers in the “amount_in_usd” column and replace the outliers with appropriate values, discussed in the sessions.

```python
Q1 = df['amount_in_usd'].quantile(0.25)
Q3 = df['amount_in_usd'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

median_value = df['amount_in_usd'].median()
```

```python
len(df[(df['amount_in_usd'] > upper_bound) | (df['amount_in_usd'] < lower_bound)])
```

```python
df['amount_in_usd'] = df['amount_in_usd'].apply(lambda x: median_value if x < lower_bound or x > upper_bound else x)
```

### 4. Try to extract new features by grouping existing columns

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.svm import LinearSVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from itertools import product
from math import sqrt
```

```python
X = df.drop('amount_in_usd', axis=1)
```

```python
y = df['amount_in_usd']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
r2_scores = {}
rmse_scores = {}
mae_scores = {}
```

```python
def run_regression_models():
    global X_train, y_train, X_test, y_test
    models = {
        "LinearRegression": LinearRegression(),
        "Ridge": Ridge(),
        "Lasso": Lasso(),
        "LinearSVR": LinearSVR(),
        "DecisionTree": DecisionTreeRegressor(),
        "RandomForest": RandomForestRegressor(),
        "AdaBoost": AdaBoostRegressor(),
        "XGBoost": XGBRegressor()
    }
    global r2_scores, rmse_scores, mae_scores
    for model_name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        r2_scores[model_name] = r2_score(y_test, y_pred)
        rmse_scores[model_name] = mean_squared_error(y_test, y_pred, squared=False)
        mae_scores[model_name] = mean_absolute_error(y_test, y_pred)
        print(f'{model_name} -> {r2_scores[model_name]} -> {rmse_scores[model_name]} -> {mae_scores[model_name]}')
```

```python
run_regression_models()
```

```python
sorted_r2 = sorted(r2_scores.items(), key=lambda x: x[1], reverse=True)
print("R2 Scores:")
for model_name, score in sorted_r2:
    print(f"{model_name}: {score}")
```

```python
sorted_rmse = sorted(rmse_scores.items(), key=lambda x: x[1])
print("\nRMSE Scores:")
for model_name, score in sorted_rmse:
    print(f"{model_name}: {score}")
```

```python
sorted_mae = sorted(mae_scores.items(), key=lambda x: x[1])
print("\nMAE Scores:")
for model_name, score in sorted_mae:
    print(f"{model_name}: {score}")
```
